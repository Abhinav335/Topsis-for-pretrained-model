```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
import time

# Function to calculate perplexity
def calculate_perplexity(model, tokenizer, text):
    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True)
    labels = tokenizer(text, return_tensors="pt", max_length=512, truncation=True)["input_ids"]
    outputs = model(**inputs, labels=labels)
    loss = outputs.loss
    perplexity = torch.exp(loss)
    return perplexity.item()

# Function to calculate diversity (using simple word overlap as an example)
def calculate_diversity(text1, text2):
    set1 = set(text1.lower().split())
    set2 = set(text2.lower().split())
    overlap = len(set1.intersection(set2))
    diversity_score = overlap / min(len(set1), len(set2))
    return diversity_score

# Function to measure training time
def measure_training_time(model, tokenizer, train_text):
    inputs = tokenizer(train_text, return_tensors="pt", max_length=512, truncation=True)
    labels = tokenizer(train_text, return_tensors="pt", max_length=512, truncation=True)["input_ids"]

    start_time = time.time()
    outputs = model(**inputs, labels=labels)
    end_time = time.time()

    training_time = end_time - start_time
    return training_time

# Replace 'EleutherAI/gpt-neo-1.3B' with the specific model identifier or URL
model_identifier = 'openai-community/gpt2'
tokenizer = GPT2Tokenizer.from_pretrained(model_identifier)
model = GPT2LMHeadModel.from_pretrained(model_identifier)

# Example text for evaluation
prompt = "Once upon a time, in a land far, far away..."

# Calculate perplexity
perplexity = calculate_perplexity(model, tokenizer, prompt)
print(f"Perplexity: {perplexity:.4f}")

# Example for diversity calculation (using a different prompt)
other_prompt = "In a galaxy not so far away, there lived..."
diversity_score = calculate_diversity(prompt, other_prompt)
print(f"Diversity Score: {diversity_score:.4f}")

# Example for measuring training time
train_text = "This is a sample training text for measuring training time."
training_time = measure_training_time(model, tokenizer, train_text)
print(f"Training Time: {training_time:.2f} seconds")
```

```python
import numpy as np

def normalize(data):
    norms = np.linalg.norm(data, axis=0)
    return data / norms

def topsis(data, weights):
    num_models, num_criteria = data.shape
    positive_ideal = np.max(data, axis=0)
    negative_ideal = np.min(data, axis=0)

    normalized_data = normalize(data)

    # Calculate distances to positive and negative ideal solutions
    d_plus = np.linalg.norm(weights * (normalized_data - positive_ideal), axis=1)
    d_minus = np.linalg.norm(weights * (normalized_data - negative_ideal), axis=1)

    # Calculate TOPSIS scores
    topsis_scores = d_minus / (d_plus + d_minus)

    return topsis_scores

# Example Data (replace with actual data)
models_data = np.array([
    [19.9, 0.33, 0.14],  # Model A
    [44.9, 0.34, 0.09],  # Model B
    [4.22, 0.34, 0.82],  # Model C
    [4.49, 0.34, 1.97]   # Model D
])

weights = np.array([0.4, 0.3, 0.3])

# Calculate TOPSIS scores
topsis_scores = topsis(models_data, weights)

# Rank the models
ranking = np.argsort(topsis_scores)[::-1] + 1  # +1 to start ranking from 1

# Print results
for i, model_rank in enumerate(ranking):
    print(f"Model {chr(65 + i)}: TOPSIS Score = {topsis_scores[i]:.4f}, Rank = {model_rank}")
```
